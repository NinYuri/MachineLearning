```python
# Importar librerías
import tensorflow as tf
import numpy as np
```

```python
# Entradas
x_Celcius = np.array([-40, -30, -20, -10, 0, 10, 20, 30, 40])
# Salidas
y_Fahrenheit = np.array([-40, -22, -4, 14, 32, 50, 68, 86, 104])

x_Celcius.shape, y_Fahrenheit.shape
```

```python
# Definir hiperparámetros
learning_rate = 0.5
epochs = 1000
```

```python
# Diseñar la red neuronal: Una capa de entrada y una de salida
# capa = tf.keras.layers.Dense(units=1,input_shape=[1])

# Crear el modelo con keras
# modelo = tf.keras.Sequential([capa])

# Crear las capas
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)

# Crear el modelo 
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
# Compilar el modelo
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate),
    loss = 'mean_squared_error'    
)
```

```python
print("Entrenando...")
historial = modelo.fit(x_Celcius, y_Fahrenheit, epochs = epochs, verbose = False)
print('Modelo Entrenado')
```

```python
# Mostrar el resultado de la función de pérdida
import matplotlib.pyplot as plt
plt.xlabel("No. Epochs")
plt.ylabel("Magnitud de Pérdida")
plt.plot(historial.history['loss'])
```

```python
# Obtener el peso y bias ajustados
print(f"Weights y bias ajustados: {salida.get_weights()}")
```

```python
# Intentar predecir a cuanto equivale 15 grados centigrados
grados_Celcius = np.array([15])
res = modelo.predict(grados_Celcius)

print(f"{grados_Celcius[0]} es equivalente a {res[0][0]} Fahrenheit")
```
