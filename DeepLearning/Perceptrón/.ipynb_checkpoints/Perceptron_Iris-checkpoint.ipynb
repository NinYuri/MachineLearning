# <font color='5D3FD3'>**PERCEPTRÓN FLORES DE LIRIO**</font>

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("\033[1;35m" + "IRIS FLOWER DATASET" + "\033[0m\n")
iris = pd.read_csv('IRIS.csv')
iris
```

#### <font color='5D3FD3'>**Operaciones al Dataset**</font>

```python
iris = iris[iris["species"] != 'Iris-virginica']

print("\033[1;35m" + "Eliminación de especie Iris-virginica" + "\033[0m\n")
iris
```

```python
iris = iris.drop('sepal_width', axis = 1)
iris = iris.drop('petal_width', axis = 1)

print("\033[1;35m" + "Eliminación de columnas innecesarias" + "\033[0m\n")
iris
```

```python
print("\033[1;35m" + "Información de nuevo dataset" + "\033[0m\n")
iris.info()
```

```python
print("\033[1;35m" + "Verificación de duplicados" + "\033[0m")
iris.duplicated().sum()
```

```python
print("\033[1;35m" + "Eliminación de filas duplicadas" + "\033[0m\n")
iris = iris.drop_duplicates()
iris
```

#### <font color='5D3FD3'>**Gráfica de Dispersión**</font>

```python
# Entradas
inputs = iris[['sepal_length', 'petal_length']].to_numpy()

# Salidas
labels = iris['species'].to_numpy()
labels = np.where(labels == 'Iris-setosa', 0, 1)
```

```python
# Gráfica de Dispersión
plt.scatter(inputs[:, 0], inputs[:, 1], c=labels, cmap='cool', marker='o', s=150)
plt.title("FLORES DE LIRIO\n", color='#a569bd', weight='bold')
plt.xlabel("\nLongitud del Sépalo", color='#c47fe0', weight='bold')
plt.ylabel("Longitud del Pétalo\n", color='#c47fe0', weight='bold')
plt.grid(True)
plt.show()
```

#### <font color='5D3FD3'>**Perceptrón**</font>

```python
# Definición de hiperparámetros
learning_rate = 0.05
bias = 0.1559
epochs = 10
weights = [0.532, 0.618]
```

```python
# Clase Perceptrón
class Perceptron():
    def __init__(self, inputs, labels, learning_rate, weights, bias, epochs = 20):
        self.inputs = inputs
        self.labels = labels
        self.learning_rate = learning_rate
        self.weights = weights
        self.bias = bias
        self.epochs = epochs

    def activation(self, z):
        return 1 if z > 0 else 0

    def predict(self, input_x, weights_fit, bias_fit):
        z = np.dot(input_x, weights_fit) + bias_fit
        y_pred = self.activation(z)
        return y_pred

    def fit(self):
        for epoch in range(self.epochs):
            print(f"\033[1;35mEpoch\033[0m {epoch + 1} / {self.epochs}", end = " ")
            total_error = 0

            for input, label in zip(self.inputs, self.labels):
                z = np.dot(input, self.weights) + self.bias
                y_pred = self.activation(z)
                
                error = label - y_pred
                total_error += abs(error)

                delta_w = self.learning_rate * error * input
                self.weights = self.weights + delta_w
                delta_bias = self.learning_rate * error
                self.bias = self.bias + delta_bias

            average_error = total_error / len(self.labels)
            print(f"\t\033[1;35mAverage Error:\033[0m {average_error}")

        return self.weights, self.bias
```

```python
# Crear el objeto perceptrón
perceptron = Perceptron(inputs, labels, learning_rate, weights, bias, epochs)

w, b = perceptron.fit()
print(f"\n\033[1;35mWeights Fit:\033[0m {w}, \033[1;35mBias Fit:\033[0m {b}")
```

#### <font color='5D3FD3'>**Predicciones**</font>

```python
print(f"\033[1;35m[4.9, 1.3] --> \033[0m{"Iris Setosa" if perceptron.predict([4.9, 1.3], w, b) == 0 else "Iris Versicolor"}")
print(f"\033[1;35m[5.7, 3.7] --> \033[0m{"Iris Setosa" if perceptron.predict([5.7, 3.7], w, b) == 0 else "Iris Versicolor"}")
print(f"\033[1;35m[5.1, 1.5] --> \033[0m{"Iris Setosa" if perceptron.predict([5.1, 1.5], w, b) == 0 else "Iris Versicolor"}")
print(f"\033[1;35m[5.4, 3.2] --> \033[0m{"Iris Setosa" if perceptron.predict([5.4, 3.2], w, b) == 0 else "Iris Versicolor"}")
print(f"\033[1;35m[6.1, 4.1] --> \033[0m{"Iris Setosa" if perceptron.predict([6.1, 4.1], w, b) == 0 else "Iris Versicolor"}")
print(f"\033[1;35m[4.8, 1.2] --> \033[0m{"Iris Setosa" if perceptron.predict([4.8, 1.2], w, b) == 0 else "Iris Versicolor"}")
```
